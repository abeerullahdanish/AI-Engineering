# Conditionals in Python

Conditionals are programming statements used to **control the flow of a program** by executing different blocks of code depending on whether a condition is **True or False**. They allow decision-making in programs.

```python
# if statement
age = 18
if age >= 18:
    print("You are adult")

# if-else
age = 16
if age >= 18:
    print("Adult")
else:
    print("Minor")

# if-elif-else
marks = 75
if marks >= 90:
    print("A Grade")
elif marks >= 70:
    print("B Grade")
elif marks >= 50:
    print("C Grade")
else:
    print("Fail")

# nested if
age = 20
citizen = True
if age >= 18:
    if citizen:
        print("Eligible to vote")

# logical operators
age = 20
student = True
if age > 18 and student:
    print("Eligible discount")

# comparison operators
x = 10
if x == 10:
    print("Equal")

````python
# 6. Write a program to calculate the grade of a student from his marks from the
# following scheme:
# 90 – 100 => Ex
# 80 – 90 => A
# 70 – 80 => B
# 60 – 70 =>C
# 50 – 60 => D
# <50 => F

marks = int(input("Enter the marks :"))
if(marks >=90 and marks <= 100):
  print("You Got Excellent Grade")
elif(marks >= 80 and marks <=90 ):
  print("You Got A Grade")
elif(marks >= 70 and marks <= 80):
  print("You Got B Grade")
elif(marks >=60 and marks <=70):
  print("You Got C Grade")
elif(marks >=50  and marks <= 60):
  print("You Got D Grade")
else:
  print("You got F Grade")
````
