# Loops in Python

Loops are used to repeat a block of code multiple times.

---

## 1. for Loop
Used when you know how many times you want to repeat.

### Syntax:
```python
for variable in sequence:
    # code

````python
for i in range(1,2):
  print(i)
````

## while Loop

Used when condition-based repetition is needed.

### Syntax:
```python
while condition:
    # code

````python
i = 1
while(i<6):
  print(i)
  i+=1

'''
output:
1
2
3
4
5
'''
````
