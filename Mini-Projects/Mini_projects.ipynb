### Project -5 Quiz Application

```python
'''Rules of a Quiz Application:
Each question has one correct answer
User can select or type only one answer per question
If answer is correct → score increases
If answer is wrong → no score is added
Each question is independent (previous question does not affect next one)
All questions must be answered (unless skip option is allowed)
Final result shows total score out of total questions (e.g., 4/5)
'''

questions = {
    "2+2" : 4,
    "Name of your country" :"pakistan",
    "2-1":0,
    "convert abeer name into uper case letter":"ABEER"
}
score = 0
for question in questions:
  print("questions :",question)

  user_answer  = input("Enter your answer").lower()
  correct_answer = questions[question]
  if user_answer == correct_answer:
    score+=1
  else:
    print("wrong answer")

print("Score",score)

```

### Project-7 Password Generator


```python
import random
import string
lenth = int(input("Enter the length of password: "))
characters = string.ascii_letters+string.digits+string.punctuation
password = " "
for i in range(lenth):
  password +=random.choice(characters)
print("Generated Password",password)
```
