### Variable Naming Rules in Python
#### Must start with a letter or underscore _, never a number.
#### Can contain letters, numbers, and underscores only — no #### spaces or special characters.
#### Case sensitive — name, Name, and NAME are all different #### variables.
#### Cannot use Python keywords like if, for, while, True, #### False, None etc.
#### No spaces allowed — use underscore instead like my_name.
#### No special characters like -, @, $, # are not allowed.
#### Can be any length — short like a or long like #### my_variable_name.
#### Best practice — use snake_case style like my_name, #### total_marks.

# Operators in Python

## 1. Arithmetic Operators

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `5 + 3` | `8` |
| `-` | Subtraction | `5 - 3` | `2` |
| `*` | Multiplication | `5 * 3` | `15` |
| `/` | Division | `5 / 3` | `1.66` |
| `//` | Floor Division | `5 // 3` | `1` |
| `%` | Modulus | `5 % 3` | `2` |
| `**` | Exponent | `5 ** 3` | `125` |

## 2. Comparison Operators

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `==` | Equal to | `5 == 5` | `True` |
| `!=` | Not equal | `5 != 3` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<` | Less than | `5 < 3` | `False` |
| `>=` | Greater or equal | `5 >= 5` | `True` |
| `<=` | Less or equal | `3 <= 5` | `True` |

## 3. Logical Operators

| Operator | Example | Result |
|----------|---------|--------|
| `and` | `5 > 3 and 2 < 4` | `True` |
| `or` | `5 > 3 or 2 > 4` | `True` |
| `not` | `not True` | `False` |

## 4. Assignment Operators

| Operator | Example | Means |
|----------|---------|-------|
| `=` | `a = 5` | Assign |
| `+=` | `a += 3` | `a = a + 3` |
| `-=` | `a -= 3` | `a = a - 3` |
| `*=` | `a *= 3` | `a = a * 3` |
| `/=` | `a /= 3` | `a = a / 3` |
| `//=` | `a //= 3` | `a = a // 3` |
| `%=` | `a %= 3` | `a = a % 3` |
| `**=` | `a **= 3` | `a = a ** 3` |

## 5. Bitwise Operators

| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `&` | AND | `5 & 3` | `1` |
| `|` | OR | `5 | 3` | `7` |
| `^` | XOR | `5 ^ 3` | `6` |
| `~` | NOT | `~5` | `-6` |
| `<<` | Left Shift | `5 << 1` | `10` |
| `>>` | Right Shift | `5 >> 1` | `2` |

## 6. Identity Operators

| Operator | Example | Meaning |
|----------|---------|---------|
| `is` | `a is b` | Same object |
| `is not` | `a is not b` | Not same object |

## 7. Membership Operators

| Operator | Example | Result |
|----------|---------|--------|
| `in` | `3 in [1,2,3]` | `True` |
| `not in` | `5 not in [1,2,3]` | `True` |
