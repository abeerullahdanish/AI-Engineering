# Types of Strings in Python

## 1. Single Line String
```python
a = "Hello World"
b = 'Hello World'
```

## 2. Multi Line String
```python

World
Python"""
```

## 3. Raw String — `r`
Backslash ko ignore karta hai.
```python
a = r"Hello\nWorld"
print(a)  # Hello\nWorld  (\n kaam nahi karta)
```

## 4. F-String — `f`
Variables directly string mein daal sakte hain.
```python
name = "Abeer"
a = f"Hello {name}"
print(a)  # Hello Abeer
```

## 5. Byte String — `b`
Network ya files ke liye use hota hai.
```python
a = b"Hello"
print(a)  # b'Hello'
```

## 6. Unicode String — `u`
Special characters ke liye.
```python
a = u"Hello \u2603"
print(a)  # Hello ☃
```

# Slicing in Python

Slicing means **extracting a part of a string** using indexes.

## Syntax
```python
string[start : stop : step]
```
- **start** — kahan se shuru karo (default 0)
- **stop** — kahan tak jao (us index ko include nahi karta)
- **step** — kitne jump karo (default 1)

## Simple Examples
```python
a = "Abeer Ullah"
#    01234567890  (index)

print(a[0:5])    # Abeer   (0 se 4 tak)
print(a[6:11])   # Ullah   (6 se 10 tak)
print(a[0:3])    # Abe     (0 se 2 tak)
```

## Skip Start or Stop
```python
a = "Abeer"

print(a[:3])     # Abe     (start se 2 tak)
print(a[2:])     # eer     (2 se end tak)
print(a[:])      # Abeer   (puri string)
```

## Negative Indexing
```python
a = "Abeer"

print(a[-3:])    # eer     (last 3 characters)
print(a[:-2])    # Abe     (last 2 chod do)
```

## Step
```python
a = "Abeer"

print(a[::2])    # Aer     (har 2nd character)
print(a[::-1])   # reebA   (string reverse)
```

## Quick Summary

| Code | Output | Matlab |
|------|--------|--------|
| `a[0:5]` | `Abeer` | 0 se 4 tak |
| `a[:3]` | `Abe` | shuru se 2 tak |
| `a[2:]` | `eer` | 2 se end tak |
| `a[::-1]` | `reebA` | reverse |
| `a[::2]` | `Aer` | har 2nd character |

# String Functions in Python

## 1. Case Functions
```python
a = "abeer ullah"

print(a.upper())       # ABEER ULLAH
print(a.lower())       # abeer ullah
print(a.title())       # Abeer Ullah
print(a.capitalize())  # Abeer ullah
print(a.swapcase())    # ABEER ULLAH
```

## 2. Search Functions
```python
a = "Abeer Ullah"

print(a.find("Ullah"))     # 6  (index return karta hai)
print(a.index("Ullah"))    # 6  (index return karta hai)
print(a.count("a"))        # 0  (kitni baar aaya)
print(a.startswith("Ab"))  # True
print(a.endswith("ah"))    # True
```

## 3. Replace & Split Functions
```python
a = "Abeer Ullah"

print(a.replace("Ullah", "Danish"))  # Abeer Danish
print(a.split())                      # ['Abeer', 'Ullah']
print(a.split("l"))                   # ['Ab', 'eer U', '', 'ah']
```

## 4. Strip Functions
```python
a = "   Abeer Ullah   "

print(a.strip())   # Abeer Ullah  (dono sides se spaces hatao)
print(a.lstrip())  # Abeer Ullah  (left side se spaces hatao)
print(a.rstrip())  # Abeer Ullah  (right side se spaces hatao)
```

## 5. Check Functions
```python
a = "Abeer123"

print(a.isalpha())   # False  (sirf letters hain?)
print(a.isdigit())   # False  (sirf numbers hain?)
print(a.isalnum())   # True   (letters ya numbers hain?)
print(a.isspace())   # False  (sirf spaces hain?)
print(a.isupper())   # False  (sab upper case?)
print(a.islower())   # False  (sab lower case?)
```

## 6. Join & Format Functions
```python
a = ["Abeer", "Ullah", "Danish"]
print(" ".join(a))   # Abeer Ullah Danish
print("-".join(a))   # Abeer-Ullah-Danish

name = "Abeer"
age = 20
print(f"My name is {name} and I am {age} years old")
```

## 7. Length Function
```python
a = "Abeer"
print(len(a))   # 5
```

# Escape Sequence Characters in Python

Escape sequences start with a **backslash** `\`

## All Escape Sequences

| Escape | Name | Example | Output |
|--------|------|---------|--------|
| `\n` | New Line | `"Hello\nWorld"` | Hello (next line) World |
| `\t` | Tab | `"Hello\tWorld"` | Hello    World |
| `\\` | Backslash | `"Hello\\World"` | Hello\World |
| `\'` | Single Quote | `'I\'m Abeer'` | I'm Abeer |
| `\"` | Double Quote | `"He said \"Hi\""` | He said "Hi" |
| `\r` | Carriage Return | `"Hello\rWorld"` | World |
| `\b` | Backspace | `"Hello\bWorld"` | HellWorld |
| `\f` | Form Feed | `"Hello\fWorld"` | Hello World |
| `\0` | Null Character | `"Hello\0World"` | Hello World |
| `\a` | Bell Sound | `"Hello\aWorld"` | Hello World (beep) |

## Examples
```python
print("Hello\nWorld")     # New line
# Hello
# World

print("Hello\tWorld")     # Tab space
# Hello    World

print("I\'m Abeer")       # Single quote
# I'm Abeer

print("He said \"Hi\"")   # Double quote
# He said "Hi"

print("Hello\\World")     # Backslash
# Hello\World
```

## Raw String — Escape Sequence Band Karo
```python
print("Hello\nWorld")    # Hello
                         # World

print(r"Hello\nWorld")   # Hello\nWorld  (\n kaam nahi karta)
```

## Quick Tip

| Use | Escape |
|-----|--------|
| Next line chahiye | `\n` |
| Tab space chahiye | `\t` |
| Quote print karni | `\'` or `\"` |
| Backslash print karni | `\\` |
