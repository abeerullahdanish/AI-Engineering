# Lists in Python

A list is a collection of **multiple values** stored in a single variable.

## Creating a List
```python
a = [1, 2, 3, 4, 5]
b = ["Abeer", "Danish", "Ullah"]
c = [1, "Abeer", True, 3.14]   # different datatypes bhi ho sakte hain
```

## Indexing
```python
a = ["Abeer", "Danish", "Ullah"]
#      0         1         2      (index)

print(a[0])    # Abeer
print(a[1])    # Danish
print(a[-1])   # Ullah  (last item)
```

## Slicing
```python
a = [1, 2, 3, 4, 5]

print(a[0:3])   # [1, 2, 3]
print(a[2:])    # [3, 4, 5]
print(a[::-1])  # [5, 4, 3, 2, 1]  (reverse)
```

## List Methods
```python
a = [1, 2, 3, 4, 5]

a.append(6)      # end mein add karo    → [1,2,3,4,5,6]
a.insert(0, 0)   # kisi bhi index pe    → [0,1,2,3,4,5,6]
a.remove(3)      # value remove karo    → [0,1,2,4,5,6]
a.pop()          # last item remove     → [0,1,2,4,5]
a.sort()         # sort karo            → [0,1,2,4,5]
a.reverse()      # reverse karo         → [5,4,2,1,0]
a.clear()        # sab remove karo      → []
print(len(a))    # length               → 0
```

## List Operations
```python
a = [1, 2, 3]
b = [4, 5, 6]

print(a + b)     # [1, 2, 3, 4, 5, 6]  (join)
print(a * 2)     # [1, 2, 3, 1, 2, 3]  (repeat)
print(3 in a)    # True                 (check)
```

## Quick Summary

| Feature | Example |
|---------|---------|
| Create | `a = [1, 2, 3]` |
| Access | `a[0]` |
| Add | `a.append(4)` |
| Remove | `a.remove(1)` |
| Length | `len(a)` |
| Sort | `a.sort()` |
| Reverse | `a.reverse()` |

````python
friends = ["Apple","orange","banana"]
print(friends[0])
friends[0]="Grapes" #unlike strigs lists are unmutable means they are changeable
print(friends)
print(friends[0::])
````

# List Methods in Python

## 1. Adding Methods
```python
a = [1, 2, 3]

a.append(4)         # end mein ek item add karo
print(a)            # [1, 2, 3, 4]

a.insert(0, 0)      # kisi bhi index pe add karo
print(a)            # [0, 1, 2, 3, 4]

a.extend([5, 6])    # doosri list ko add karo
print(a)            # [0, 1, 2, 3, 4, 5, 6]
```

## 2. Removing Methods
```python
a = [1, 2, 3, 4, 5]

a.remove(3)         # value se remove karo
print(a)            # [1, 2, 4, 5]

a.pop()             # last item remove karo
print(a)            # [1, 2, 4]

a.pop(0)            # index se remove karo
print(a)            # [2, 4]

a.clear()           # sab remove karo
print(a)            # []
```

## 3. Searching Methods
```python
a = [1, 2, 3, 2, 1]

print(a.index(2))   # 1      (pehli baar kahan hai)
print(a.count(2))   # 2      (kitni baar hai)
print(len(a))       # 5      (total items)
```

## 4. Sorting Methods
```python
a = [3, 1, 4, 1, 5]

a.sort()            # ascending order
print(a)            # [1, 1, 3, 4, 5]

a.sort(reverse=True) # descending order
print(a)            # [5, 4, 3, 1, 1]

a.reverse()         # reverse karo
print(a)            # [1, 1, 3, 4, 5]
```

## 5. Copying Methods
```python
a = [1, 2, 3]

b = a.copy()        # copy banao
print(b)            # [1, 2, 3]
```

## Quick Summary

| Method | Kaam |
|--------|------|
| `append()` | end mein add |
| `insert()` | kisi index pe add |
| `extend()` | doosri list add |
| `remove()` | value se remove |
| `pop()` | index se remove |
| `clear()` | sab remove |
| `index()` | index dhundo |
| `count()` | count karo |
| `sort()` | sort karo |
| `reverse()` | reverse karo |
| `copy()` | copy banao |
| `len()` | length pato |

# Tuples in Python

A tuple is just like a list but **cannot be changed** after creation.

## Creating a Tuple
```python
a = (1, 2, 3, 4, 5)
b = ("Abeer", "Danish", "Ullah")
c = (1, "Abeer", True, 3.14)   # different datatypes bhi ho sakte hain
d = (1,)                        # single item tuple — comma zaruri hai!
```

## Indexing
```python
a = ("Abeer", "Danish", "Ullah")
#      0         1         2      (index)

print(a[0])    # Abeer
print(a[1])    # Danish
print(a[-1])   # Ullah  (last item)
```

## Slicing
```python
a = (1, 2, 3, 4, 5)

print(a[0:3])   # (1, 2, 3)
print(a[2:])    # (3, 4, 5)
print(a[::-1])  # (5, 4, 3, 2, 1)  (reverse)
```

## Tuple Methods
```python
a = (1, 2, 3, 2, 1)

print(a.index(2))   # 1   (pehli baar kahan hai)
print(a.count(2))   # 2   (kitni baar hai)
print(len(a))       # 5   (total items)
```

## Tuple Operations
```python
a = (1, 2, 3)
b = (4, 5, 6)

print(a + b)     # (1, 2, 3, 4, 5, 6)  (join)
print(a * 2)     # (1, 2, 3, 1, 2, 3)  (repeat)
print(3 in a)    # True                 (check)
```

## Tuple vs List

| Feature | List | Tuple |
|---------|------|-------|
| Brackets | `[ ]` | `( )` |
| Changeable | ✅ Yes | ❌ No |
| Methods | Many | Only 2 |
| Speed | Slow | Fast |
| Use | Data changes | Data fixed |

## Why Use Tuple?
- **Faster** than list
- Data **safe** rahta hai — koi change nahi kar sakta
- Use karo jab data **fix** ho jaise days of week, months

```python

months = ("Jan", "Feb", "Mar", "Apr", "May", "Jun")
```

# Tuple Methods in Python

Tuple mein sirf **2 methods** hote hain kyunki ye changeable nahi hota.

## 1. index()
Pehli baar kahan hai wo index return karta hai.
```python
a = (1, 2, 3, 2, 1)

print(a.index(2))    # 1  (2 pehli baar index 1 pe hai)
print(a.index(1))    # 0  (1 pehli baar index 0 pe hai)
```

## 2. count()
Kitni baar aaya wo count karta hai.
```python
a = (1, 2, 3, 2, 1)

print(a.count(2))    # 2  (2 do baar aaya)
print(a.count(1))    # 2  (1 do baar aaya)
print(a.count(3))    # 1  (3 ek baar aaya)
```

## Other Useful Functions

### len()
```python
a = (1, 2, 3, 4, 5)
print(len(a))        # 5
```

### in operator
```python
a = (1, 2, 3, 4, 5)
print(3 in a)        # True
print(6 in a)        # False
```

### min() and max()
```python
a = (1, 2, 3, 4, 5)
print(min(a))        # 1
print(max(a))        # 5
```

### sum()
```python
a = (1, 2, 3, 4, 5)
print(sum(a))        # 15
```

### sorted()
```python
a = (3, 1, 4, 1, 5)
print(sorted(a))     # [1, 1, 3, 4, 5]  (list return karta hai)
```
