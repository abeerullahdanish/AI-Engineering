# Introduction to Dictionary

A **Dictionary** stores data in **key : value** pairs. It is like a real-life dictionary where you look up a "word" (key) to find its "meaning" (value).

### Key Points:
*   **Syntax:** Written in curly braces `{ }`.
*   **Structure:** Every key is connected to its value by a colon `:`.
*   **Unique Keys:** Keys cannot be duplicated.
*   **Mutable:** You can change, add, or remove items after the dictionary is created.
*   **Ordered:** Since Python 3.7, dictionaries maintain the order in which items are inserted.

### Simple Example:
```python
student = {
    "name": "Abeer",
    "age": 23,
    "city": "Lahore"
}

print(student)
```

### Breakdown:
*   `"name"` $\rightarrow$ **Key**
*   `"Abeer"` $\rightarrow$ **Value**


````python
info = {
    "name" : "Abeer Ullah Danish",
    "Age" : 23,
    "marks" : 677
}
print(info)
print(info["name"])
# print(info[0])
````

````python
info = {
    "name" : "Abeer Ullah Danish",
    "Age" : 23,
    "marks" : 677
}
print(info.items())
print(info.keys())
info.update({"name":"Abeer","marks":1065 })
print(info )
print(info.get("name"))
````

# Introduction to Sets

A **Set** is an unordered collection of items where every element is **unique**. It is mostly used when you want to remove duplicates or perform mathematical operations like unions and intersections.

### Key Points:
*   **No Duplicates:** A set automatically removes any repeated items.
*   **Unordered:** Items do not have a fixed position (index), so you cannot access them by a number like `my_set[0]`.
*   **Curly Braces `{ }`:** Sets use curly braces, but unlike dictionaries, they don't have `key:value` pairs.
*   **Mutable:** You can add or remove items, but the individual items themselves must be of a type that cannot be changed (like strings or numbers).

### Simple Example:
```python
# Even if we put duplicates, Python removes them
my_set = {1, 2, 2, 3, 4, 4}

print(my_set)
# Output: {1, 2, 3, 4}
```

### Quick Comparison:
*   **List:** `[1, 2, 2]` (Ordered & allows duplicates)
*   **Set:** `{1, 2}` (Unordered & unique only)


````python
s = {1,5,32}
# print(s)
e = set() # this is used ofr emptypy set
print(type(e))
print(len(s))
s.remove(5)
print(s)
s.pop()
print(s)
````

````python
s1 = {1,45,6}
s2 = [7,8,1,78]
print(s1.union(s2))
print(s1.intersection(s2))
````
